# Presentation Plots

```python
from __future__ import print_function, division
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
%config InlineBackend.figure_format = 'png'
import matplotlib as mpl
mpl.rcParams['figure.dpi']= 300
```

```python
test = [0.5461,0.6253,0.6449,0.6988,0.7469,0.7506,0.7875]
test_type = ['Dummy','SVC','Decision Tree','Random Forest','Random Forest 2','Random Forest 3','Random Forest 4']

xs = range(len(test_type))
ys = [.8285] * len(test)
```

```python
# Make the plot
plt.figure(num=None, figsize=(8, 4), dpi=300, facecolor='w', edgecolor='k')
plt.plot(xs,test,'o-',
         color='#0FB4BB', label='My Model')
plt.plot(xs,ys, color = '#F05C6A', label='Current Leader')
plt.xlabel("\nModel Type",fontsize=15)
plt.ylabel("Test Score",fontsize=15)
plt.title("How do I compare?\n", fontsize=17)
plt.xticks(xs,test_type)
plt.xticks(rotation=25)
plt.ylim((0.4,1))
plt.legend(loc="lower left")
plt.show()
```
