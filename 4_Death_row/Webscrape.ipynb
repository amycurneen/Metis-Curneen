```python
import urllib.request
from bs4 import BeautifulSoup
import csv
import time
```

```python
def run():
    #  get the webpage with the prisoner data
    uri = 'http://www.tdcj.state.tx.us/death_row/dr_executed_offenders.html'

    f = csv.writer(open("death_row_data_extended_onlyfull" + ".csv", "w", newline=''), dialect='excel')

    # Write column headers as the first line
    f.writerow(['execution', 'last_statement', 'last_name', 'first_name', 'TDCJ', 'age_execution', 
                'execution_date', 'date_of_birth', 'date_received', 'age_received', 'education_level', 
                'date_of_offense', 'age_offense', 'county', 'race', 'gender', 'hair_color', 'height', 
                'weight', 'eye_color', 'native_county', 'native_state', 'prior_occupation', 
                'prior_prison_record', 'summary_of_incident', 'co_defendants', 'race_and_gender_of_victim'])

    urllines = urllib.request.urlopen(uri)
    pagedat = urllines.read()
    urllines.close()
    soup = BeautifulSoup(pagedat)
    allrows = soup.find_all("tr")
    suburi = uri[:38]
    
    for row in allrows:
        tds = row.find_all("td")
        try:
            #links
            links = row.find_all('a')
            
            # offenders link
            link = links[0].get('href')
            OLink = suburi+link
            
            #last statement link
            link = links[1].get('href')
            LSLink = suburi+link
            
            #last statement page
            
            urllines2 = urllib.request.urlopen(LSLink)
            pagedat2 = urllines2.read()
            urllines2.close()
            soup2 = BeautifulSoup(pagedat2)
            
            par = soup2.find_all("p")
            for i in range(1, (len(par)-1)):
                if 'Last Statement:' in par[i].get_text():
                    last_statement = str(par[i+1].get_text())
                    last_statement = last_statement.replace('\x92s', '')
                    last_statement = last_statement.replace('\xa0', '')
            
            if last_statement == '':
                last_statement = 'No Statement'
                    
            #offenders page
            
#             if OLink[-4:] == '.jpg':   
#                 date_of_birth = ''
#                 date_received = ''
#                 age_received = ''
#                 education_level = ''
#                 date_of_offense = ''
#                 age_offense = ''
#                 county = ''
#                 race = ''
#                 gender = ''
#                 hair_color = ''
#                 height = ''
#                 weight = ''
#                 eye_color = ''
#                 native_county = ''
#                 native_state = ''
#                 prior_occupation = ''
#                 prior_prison_record = ''
#                 summary_of_incident = ''
#                 race_and_gender_of_victim = ''
                
#             else:    
            urllines1 = urllib.request.urlopen(OLink)
            pagedat1 = urllines1.read()
            urllines1.close()
            soup1 = BeautifulSoup(pagedat1)

            tds_off = soup1.find_all("td")   
            date_of_birth = str(tds_off[6].get_text())
            date_received = str(tds_off[8].get_text())
            age_received = str(tds_off[10].get_text())
            education_level = str(tds_off[12].get_text())
            date_of_offense = str(tds_off[14].get_text())
            age_offense = str(tds_off[17].get_text())
            #county = str(tds_off[20].get_text())
            #race = str(tds_off[23].get_text())
            gender = str(tds_off[26].get_text())
            hair_color = str(tds_off[29].get_text())
            height = str(tds_off[32].get_text())
            weight = str(tds_off[35].get_text())
            eye_color = str(tds_off[38].get_text())
            native_county = str(tds_off[41].get_text())
            native_state = str(tds_off[44].get_text())

            par = soup1.find_all("p")
            for i in range(0, (len(par))):
                if 'Prior Occupation' in par[i].get_text():
                    prior_occupation = str(par[i].get_text())
                    prior_occupation = prior_occupation[18:]
                elif 'Prior Prison Record' in par[i].get_text():
                    prior_prison_record = str(par[i].get_text())
                    prior_prison_record = prior_prison_record[23:]
                elif 'Summary of Incident' in par[i].get_text():
                    summary_of_incident = str(par[i].get_text())
                    summary_of_incident = summary_of_incident[24:]
                elif 'Co-Defendants' in par[i].get_text():
                    co_defendants = str(par[i].get_text())
                    co_defendants = co_defendants[17:]
                elif 'Race and Gender of Victim' in par[i].get_text():
                    race_and_gender_of_victim = str(par[i].get_text())
                    race_and_gender_of_victim = race_and_gender_of_victim[29:]

            #main page
            
            execution = str(tds[0].get_text())
            last_name = str(tds[3].get_text())
            first_name = str(tds[4].get_text())
            TDCJ = str(tds[5].get_text())
            age_execution = str(tds[6].get_text())
            execution_date = str(tds[7].get_text())
            race = str(tds[8].get_text())
            county = str(tds[9].get_text())
            
            print("good string")
            
        except Exception:
            print("bad string")
            time.sleep(.1)
            continue
            
        f.writerow([execution, last_statement, last_name, first_name, TDCJ, age_execution, execution_date, 
        date_of_birth, date_received, age_received, education_level, date_of_offense, age_offense, 
        county, race, gender, hair_color,height, weight, eye_color, native_county, native_state,
        prior_occupation, prior_prison_record, summary_of_incident, co_defendants, race_and_gender_of_victim])
        
        last_statement = ''

if __name__ == "__main__":
    run()
    print("job complete!")
```
