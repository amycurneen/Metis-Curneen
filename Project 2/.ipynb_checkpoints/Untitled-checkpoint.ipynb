```python
name = 'Donald Knuth'
name = name.replace(' ','').lower()
```

```python
name
```

```python
name_set = []

for i in range(len(name)):
    name_set.append(name[i])
    for j in range(len(name)-i-1):
        two = name[i] + name[j+i+1]
        name_set.append(two)
    
name_set
    
    
    
    
```

```python
combs = []

import itertools

for r in range(1,len(name)+1):
    els = [list(x) for x in itertools.combinations(name, r)]
    combs.extend(els)
combs
```

```python
l2 = [''.join(e) for e in combs]

l2
```
